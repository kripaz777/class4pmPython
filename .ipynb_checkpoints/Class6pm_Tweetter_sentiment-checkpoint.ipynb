```python
import pandas as pd #we import pandas library to be able to work on dataframes in the csv file 
df = pd.read_csv('/content/covid_tweets.csv', usecols=['Tweets']) 
df
```

```python
data = df
new_data = data.drop_duplicates(subset="Tweets") #to drop retweets and repeated rows #subset is a keyword in pandas with column value as "Tweets" to indicate the column with duplicate text
new_data.to_csv('/content/filtered_data.csv')
new_data
```

```python
# Text pre-processing 
def get_normalize(text):
    text = re.sub(r'\&\w*;','',text)
    text = re.sub('@[^\s]+','',text)
    text = re.sub(r'\$\w*','',text)
    text = text.lower()
    text = re.sub("[^a-zA-Z]", " ", text) 
    text = re.sub(r'https?:\/\/.*\/\w*','',text)
    text = re.sub(r'#\w*','',text)
    text = re.sub(r'[' + string.punctuation.replace('@','') + ']+',' ',text)
    text = re.sub(r'\b\w{1,2}\b','',text)
    text = re.sub(r'\s\s+',' ',text)
    text = text.lstrip(' ')
    text = ''.join(c for c in text if c <= '\uFFFF')
    return text
j = 0
for i in new_data['Tweets']:
  new_data.loc[j]=get_normalize(str(i))
  j = j+1
new_data
```

```python
from textblob import TextBlob

new_data[['Polarity', 'Subjectivity']] = new_data['Tweets'].apply(
    lambda Text: pd.Series(TextBlob(Text).sentiment))
new_data
```

```python
new_data.to_csv('/content/filtered_data.csv')
```

```python
y = new_data['Polarity']
positive = 0
negative = 0
neutral = 0

for a in y:
    if a>0:
     positive=positive+1
    elif a<0:
     negative=negative+1
    else:
     neutral=neutral+1
numberOfTweetsAnalysed= positive+negative+neutral
print('Number of Tweets Analysed = '+str(numberOfTweetsAnalysed))
print('Positive tweets = '+str(positive))
print('Negative tweets = '+str(negative))
print('Neutral tweets = '+str(neutral))

```

```python
import csv
b = open('/content/filtered_data.csv','r')
reader = csv.DictReader(b)
for i in reader:
  print(i)
  final_data = []

  try:
    x = dict(i)
    if x['Polarity'] > 0:
      x['labels'] = "Positive"
    elif x['Polarity'] == 0:
      x['labels'] = "Neutral"
    else:
      x['labels'] = "Negative"
    final_data.append(x)
  except:
    pass

print(final_data)
```

```python
import csv
csv_columns = ['', 'Tweets','Subjectivity', 'Polarity', 'labels']
dict = final_data
try:
    with open("new_tweets.csv", 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in dict:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
